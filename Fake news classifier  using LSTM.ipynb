```python
import pandas as pd
```

```python
df = df.dropna()
```

```python
# Independent Variable
X = df.drop(['label'], axis =1)
```

```python
# Dependent Variable
y = df['label']
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
import numpy as np
print(np.__version__)

import keras
print(keras.__version__)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
# Vocabulary Size
voc_size = 10000    #It represents the top 10000 most frequent words present in dataset
```

```python
# OneHot Representation
message = X.copy()
message['title']
```

```python
message = message.reset_index(drop=True)
```

```python
message['title'][1]
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# DataSet Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus=[]

for i in range(0 ,len(message)):
  review = re.sub('[^a-zA-Z]', ' ',message['title'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus [1]
```

```python
# One Hot Representation
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr[1]
```

```python
corpus[1]
```

```python
# Finding the longest Sent
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

lengths = [len(word_tokenize(sent)) for sent in corpus]

max_index = lengths.index(max(lengths))

print(f"Longest sentence: {corpus[max_index]}")
print(f"Length of longest sentence: {lengths[max_index]}")
```

```python
# Embedding Representation

sent_length = 50

embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen = sent_length)

print(embedded_docs)
```

```python
embedded_docs[1]
```

```python
# Creating Model
embedding_vector_feature = 100
model = Sequential()
model.add(Embedding(input_dim = voc_size,output_dim = embedding_vector_feature, input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Build the model by passing data through
model.build(input_shape=(None, sent_length))
```

```python
print(model.summary())
```

```python
len(embedded_docs), y.shape
```

```python
import numpy as np
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape , y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.33, random_state = 42)
```

```python
# MOdel Training

model.fit(X_train,y_train, validation_data=(X_test, y_test), epochs=10, batch_size = 64)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred > 0.5, 1, 0)   #AUC ROC Curve
```

```python
from sklearn.metrics import accuracy_score

print(f"Accuracy Score: {accuracy_score(y_test,y_pred)}")
```

```python
from sklearn.metrics import classification_report
print(f"Classification Report : \n{classification_report(y_test, y_pred)}")
```

```python
from sklearn.metrics import confusion_matrix
print(f"Confusion Matrix : \n{confusion_matrix(y_test, y_pred)}")
```

```python
# Adding the Dropout Layer in the Model Building
from tensorflow.keras.layers import Dropout

embedding_vector_feature = 200
model = Sequential()
model.add(Embedding(input_dim = voc_size, output_dim = embedding_vector_feature))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer ='adam',metrics=['accuracy'])
model.build(input_shape = (None, sent_length))
```

```python
print(model.summary())
```

```python
model.fit(X_train, y_train,validation_data=(X_test, y_test), epochs=10, batch_size = 64)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = np.where(y_pred > 0.5, 1,0)
```

```python
print(f"Accuracy Score : \n{accuracy_score(y_pred, y_test)}")
print(f"Classification Report : \n{classification_report(y_pred, y_test)}")
print(f"Confusion matrix : \n{confusion_matrix(y_pred, y_test)}")
```
